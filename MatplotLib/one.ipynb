```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.arange(1,10)
y=x**2
```

```python
y
```

# Basic Example

```python
plt.plot(x,y,'r')
plt.xlabel('X Axis Title Here')
plt.ylabel('Y Axis Title Here')
plt.title('String Title Here')
plt.show()
```

```python
# plt.subplot(nrows, ncols, plot_number)
plt.subplot(1,3,1)
plt.plot(x, y, 'r--') # More on color options later
plt.subplot(1,2,2)
plt.plot(y, x, 'g*-')
plt.show()
```
